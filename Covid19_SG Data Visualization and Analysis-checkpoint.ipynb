## Data Manipulation

```python
# Fill in all missing values with 0 and print out the first 5 rows
df_no_missing = df.fillna(0)
df_no_missing.head()
```

```python
# Visualizing dataset after filling all NaN/null values
df_no_missing.isnull().sum().plot(kind = "barh")
plt.show()
```

```python
# Examining the days where False Positives cases are found 
df_no_missing[df_no_missing["False Positives Found"] != 0]["False Positives Found"].value_counts()
```

```python
mapping = {1.0 : True, 41.0 : True, 35.0 : True, 0.0 : False}
df_no_missing["False Positives Rates"] = df_no_missing["False Positives Found"].map(mapping)
df_no_missing["False Positives Rates"].mean()
```

```python
sns.countplot(df_no_missing["False Positives Rates"])
plt.show()
```

### False Positive cases are recorded around 0.138% of the times. PCR and SWAB tests results were largely accurate

## To examine the severity of coronavirus pandemic. <br></br> Dataset is split among 2 sections namely : Pre-Circuit Breaker, Circuit Breaker.

### Pre-Circuit Breaker

```python
# Split up data_no_missing dataset: Pre-Circuit Breaker
# We are only going to examine 8 columns here
df_pre_cb = df_no_missing.set_index("Date").sort_index(ascending = True).loc["2020-01" : "2020-04-06"]
severity_col = ["Daily Confirmed ", "Cumulative Confirmed", "Daily Discharged", "Cumulative Discharged", "Discharged to Isolation", "Still Hospitalised", "Daily Deaths", "Cumulative Deaths"]
df_pre_cb = df_pre_cb.loc[: , severity_col]
df_pre_cb
```

```python
# List of matplotlib built-in styles available
print(plt.style.available)
```

```python
# Next we want to visually examine why is our nation in need of circuit breaker.
# Plotting a line graph 
plt.style.use("default")
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(df_pre_cb.index, df_pre_cb["Daily Confirmed "], color = "g", alpha = 0.5)
ax.set_xlabel("Date (YYYYMMDD)")
ax.set_ylabel("Daily Confirmed Cases", color = "g")
ax.tick_params("y", color = "g")
ax2 = ax.twinx()
ax2.plot(df_pre_cb.index, df_pre_cb["Cumulative Confirmed"], color = "r", alpha = 0.5)
ax2.set_ylabel("Cumulative Confirmed Cases", color = "r")
ax2.tick_params("y", color = "r")
plt.show()
```

```python
#Visually examining possible correlation between number of Still Hospitalized and Daily confirmed cases. 
sns.lmplot(x = "Daily Confirmed " ,y = "Still Hospitalised", data = df_pre_cb, fit_reg = False)
plt.show()
```

```python
#Visually examining possible correlation between number of cumulative discharged and Still Hospitalised cases. 
sns.lmplot(x = "Cumulative Discharged" ,y = "Still Hospitalised", data = df_pre_cb)
plt.show()
```

```python
# Total number of covid-19 related deaths before circuit breaker started
df_pre_cb["Cumulative Deaths"].max()
```

### Despite having very low covid-19 related death cases. Daily confirmed cases have accelerated and resulted in sharp cumulative confirmed cases increment since the first covid-19 case. An steep increment in cumulative confirmed has been consistenly adding covid-19 patients in need of hospitalization care. The overall rate of cumulative discharged is lesser than confirmed covid-19 patients who required hospitalization care <br>
### This puts a heavy burden on hospital healthcare services, hospital wards and beds.

### Moving into Circuit Breaker

```python
# Split up data_no_missing dataset : Circuit Breaker

df_cb = df_no_missing.set_index("Date").sort_index(ascending = True)
df_cb = df_cb.loc["2020-04-07" : "2020-06-01"]
df_cb = df_cb.loc[:, severity_col]
df_cb[df_cb["Cumulative Confirmed"] > 10000];
```

```python
# Subsetting highest daily confirmed row 

cb_daily_max = df_cb["Daily Confirmed "].max()
df_cb[df_cb["Daily Confirmed "] == cb_daily_max]
```

```python
df_cb.loc["2020-04-07" : "2020-04-20", "Daily Confirmed "].plot(color = "r")
plt.ylabel("Daily Confirmed Cases")
plt.xlabel("Date - Between start of circuit breaker to daily confirmed cases peaked")
plt.show()
```

```python
# Visualizing highest daily confirmed case during Circuit Breaker
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(df_cb.index , df_cb["Daily Confirmed "], color = "r")
ax.set_xlabel("Date (YYYYMMDD)")
ax.set_ylabel("Daily Confirmed Cases during circuit breaker")
ax.set_title("Daily Confirmed Cases (During Circuit Breaker) over time")
ax.annotate("2020-04-20", xy = (pd.Timestamp("2020-04-20"), 1426), xytext = [pd.Timestamp("2020-04-22"), 400], arrowprops = {"arrowstyle" : "->", "color" : "blue", "lw" : 5,"alpha" : 100})
plt.show()
```

### Above plot illustrates the steep inclination of number of new daily confirmed cases reported since circuit breaker started. The increment lasted 12 days before tapering off gently over the next few weeks

```python
df_cb_pie = df_cb[df_cb["Daily Confirmed "] == df_cb["Daily Confirmed "].max()]
df_cb_pie.iloc[0,1], df_cb_pie.iloc[0,2] 
```
